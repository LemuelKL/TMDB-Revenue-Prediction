```python
import pandas as pd
```

```python
# Example DataFrame
data = {
    "categories": [["A", "B"], ["B", "C"], ["A", "C"], ["B", "D"]],
    "target": [10, 15, 12, 18],  # Example target variable
}
df = pd.DataFrame(data)

# Flatten the 'categories' column
flattened_categories = [
    category for sublist in df["categories"] for category in sublist
]

# Create a new DataFrame with the flattened categories and corresponding target values
category_target_df = pd.DataFrame(
    {
        "categories": flattened_categories,
        "target": df["target"].repeat(df["categories"].apply(len)),
    }
)

# Calculate target mean encoding for each category
target_means = category_target_df.groupby("categories")["target"].mean()

# Map the target means to the original 'categories' column in the main DataFrame
df["encoded_categories"] = df["categories"].apply(
    lambda x: [target_means[cat] for cat in x]
)

# Display the resulting DataFrame with target encoded categories
print(df)
```
